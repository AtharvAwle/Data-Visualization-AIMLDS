```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Categorical Data Plots
'''
boxplot 
violinplot
stripplot
swarmplot
barplot
countplot
'''
```

```python
dataset = sns.load_dataset('tips')
dataset
```

```python
# COUNT PLOT - used to categorically count the columns bases on category

sns.countplot(x = dataset['sex'])
```

```python
# If we have to check how many male and female smoke or compare any other category we will use hue and pass another column init
sns.countplot(x = dataset['sex'],hue = dataset['smoker'])
```

```python
# Barplot

sns.barplot(x = dataset['sex'], y = dataset['total_bill'])
# in barplot x --> qualitative.  & y --> quantative
```

```python
# bar plots also have estimator wherein we can change how quantative axis is measured by mean median sum count

sns.barplot(x = dataset['sex'] , y = dataset['total_bill'] , estimator= np.sum)
```
