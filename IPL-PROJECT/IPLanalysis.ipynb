```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("IPL.csv")
data.head()
```

### Which team won the most matches?

```python
matchwins = data['match_winner'].value_counts()
```

```python
sns.barplot(y = matchwins.index , x = matchwins , palette="mako")
plt.title("Number of Matches Won by each Team", fontsize=16)
plt.xlabel("No.of matches")
plt.ylabel("Teams")
```

### Toss decision Trends

```python
data['toss_decision'].value_counts()
```

```python
sns.countplot(x = data['toss_decision'])
plt.title("Toss decision Trends and frequency")
plt.xlabel("Toss Decision")
plt.ylabel("Frequency")
```
