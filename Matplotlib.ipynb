```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

** Basic Matplotlib Plots **

```python
xaxis = np.linspace(0,5,11)
xaxis
yaxis = xaxis ** 2
yaxis
```

```python
# Plotting basic line graph with lables and titles

plt.plot(xaxis , yaxis)
plt.title("Basic Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

** Sub Plots in matplot lib **

```python
# Making multiple plots together

plt.subplot(2,2,1)
plt.plot(xaxis,yaxis)
plt.subplot(2,2,2)
plt.plot(xaxis,xaxis)
plt.subplot(2,2,3)
plt.plot(xaxis ** 2 , xaxis)
plt.subplot(2,2,4)
plt.plot(yaxis,xaxis)
```

** Object Oriented Way **

```python
fig = plt.figure(figsize=(10,5), dpi= 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(xaxis,yaxis)
axis1.set_title("Axis 1")
axis1.set_xlabel("X")
axis1.set_ylabel("Y")


axis2 = fig.add_axes([0.2,0.8,0.2,0.2])
axis2.plot(xaxis,yaxis)
axis2.set_title("Axis 2")

# we can play around with out canvas and place the plots where every we like
```
