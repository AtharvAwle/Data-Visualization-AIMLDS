```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# We use distribution plot to visualize the distribution of quantitative data.
```

```python
# Seaborn already has some inbuild datasets for practice
dataset = sns.load_dataset('tips') 
dataset
```

```python
# here in this dataset only total_bill and tip are quantitative variables 
```

```python
# 1 - histplot()
plt.subplot(1,2,1)
sns.histplot(dataset['total_bill'],kde = True)

plt.subplot(1,2,2)
sns.histplot(dataset['tip'],kde = True)
```

```python
# 2 - jointplot()
# If we have to keep the values of x and y different we can use jointplot()

sns.jointplot(x = 'total_bill' , y = 'tip' , data = dataset ,kind = 'scatter')
#            (x axis value.    , y axis value ,  data = where is your data stored,   kind = in what kind of fashion you want your data to be plot)
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = dataset , kind = 'reg') #kind = kde,hex,reg
```

### 3 - pair plot (IMP)

```python
# it directly plots pairwise relationships in a dataset using all quantitative variables.
sns.pairplot(dataset,hue = 'sex')
#.                   hue = any categorical variable present in the dataset(qualitative data)
```

```python
# 4 - rugplot()

sns.rugplot(dataset)
```
